```python
from scipy.stats import chi2_contingency
 
# defining the table
data = [[207, 282, 241], [234, 242, 232]]
stat, p, dof, expected = chi2_contingency(data)
 
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
for _column in [c for c in obj_df.columns if c != 'Attrition']:
    _cat_data = pd.crosstab(obj_df[_column], obj_df['Attrition'])
    stat, p, dof, expected = scipy.stats.chi2_contingency(_cat_data.values)
    display( 
                _cat_data
           )
    print('persuambly dependet' if p <= .05 else 'independent')
```

Numerical features

```python
num_df = df.select_dtypes(include=[np.number])
```

```python
fig, axes = plt.subplots(12, 2, figsize=(12, 22)) 

for i, feature in enumerate(num_df.columns):
    row, col = divmod(i, 2)  
    sns.kdeplot(num_df[feature], ax=axes[row, col], fill=True)
    axes[row, col].set_title(f'{feature} Distribution', size=11)
    axes[row, col].set_xlabel('') 
    axes[row, col].set_ylabel('') 
    
plt.tight_layout()
```

```python
plt.figure(figsize=(15,7))
sns.kdeplot(data=df, x = "DailyRate", hue="Attrition")
```

```python
# correlation
corr_features_set = pd.concat([df.select_dtypes(include=[np.number]),(df.Attrition == 'Yes').astype(int)], axis=1)
corr = corr_features_set.corr()

# set the size of the plot
fig, ax = plt.subplots(figsize = (16, 8))

# Generate a mask for upper traingle
mask = np.triu(np.ones_like(corr, dtype = bool))

# Configure a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap = True)

# Draw the heatmap
sns.heatmap(corr, annot = True, mask = mask, cmap = cmap)
```
